# Smoker-Status_prediction

## Импорт библиотек

```python
# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn import tree
```

## Загрузка данных

```python
# Загружаем данные
df = pd.read_csv('data.csv')  # Укажи путь к своему файлу

# Посмотрим на первые несколько строк данных, чтобы понять структуру
df.head()
```

## Исследовательский анализ данных (EDA)

```python
# Информация о данных
df.info()
```

```python
# Статистические показатели
df.describe()
```

```python
# Проверка на пропуски
df.isnull().sum()
```

## Визуализация распределения целевой переменной



```python
# Визуализируем распределение целевой переменной (статус курения)
sns.countplot(x='smoking', data=df)
plt.title('Распределение статуса курения')
plt.xlabel('Статус курения')
plt.ylabel('Количество')
plt.show()
```

```python
# Преобразуем категориальные переменные в числовые (если есть)
# Пример: преобразуем переменные зрения и слуха в числовые значения (если они категориальные)
df['eyesight(left)'] = pd.to_numeric(df['eyesight(left)'], errors='coerce')
df['eyesight(right)'] = pd.to_numeric(df['eyesight(right)'], errors='coerce')
df['hearing(left)'] = pd.to_numeric(df['hearing(left)'], errors='coerce')
df['hearing(right)'] = pd.to_numeric(df['hearing(right)'], errors='coerce')

# Проверим изменения
df[['eyesight(left)', 'eyesight(right)', 'hearing(left)', 'hearing(right)']].head()
```

## Заполнение пропусков

```python
# Заполнение пропусков средними значениями для числовых переменных
df.fillna(df.mean(), inplace=True)

# Проверим, остались ли пропуски
df.isnull().sum()
```

## Разделение данных на признаки и целевую переменную

```python
# Разделяем данные на признаки (X) и целевую переменную (y)
X = df.drop(['id', 'smoking'], axis=1)  # Убираем 'id' и 'smoking'
y = df['smoking']  # Целевая переменная (статус курения)

# Разделим данные на обучающую и тестовую выборки (70% обучающая, 30% тестовая)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Размеры выборок
X_train.shape, X_test.shape
```

## Построение модели и обучение

```python
# Создаем и обучаем модель Decision Tree
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Предсказания на тестовых данных
y_pred = model.predict(X_test)
```

## Оценка модели

```python
# Оценка модели
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Классификационный отчет
print(classification_report(y_test, y_pred))

# Матрица ошибок
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Не курит', 'Курит'], yticklabels=['Не курит', 'Курит'])
plt.title('Матрица ошибок')
plt.xlabel('Предсказания')
plt.ylabel('Реальные значения')
plt.show()
```

## Важность признаков

```python
# Важность признаков для модели Decision Tree
feature_importance = model.feature_importances_
features = X.columns

# Строим график важности признаков
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance, y=features)
plt.title('Важность признаков')
plt.xlabel('Важность')
plt.ylabel('Признаки')
plt.show()
```
