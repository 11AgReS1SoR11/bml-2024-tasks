### Обработку коррелирующих признаков в данных

Этот пункт является важным, поскольку большая корреляция признаков мешает моделям работу в виду увеличения значимости тех или иных данных и переобучения модели на них.

```python
sns.heatmap(X[num_feat].corr())
```

```python
# Функция фильтрации фичей по корреляции
def feats_filter_corr(df: pd.DataFrame,
                      features: list,
                      corr_threshold: float):
    correlation_matrix = df[num_feat].corr()
    for_del_feats = set()
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if (abs(correlation_matrix.iloc[i, j]) >= corr_threshold):
                if (df[correlation_matrix.columns[i]].isna().mean() < df[correlation_matrix.columns[j]].isna().mean()):
                    for_del_feats.add(correlation_matrix.columns[i])
                else:
                    for_del_feats.add(correlation_matrix.columns[j])

    return [feat for feat in features if feat not in for_del_feats]

features = feats_filter_corr(X, features, 0.95)
num_feat = [feat for feat in num_feat if feat in features]
cat_feat = [feat for feat in cat_feat if feat in features]
len(features)
```

Мы пытались отсеять по трешхолду 95% (то есть убирать те признаки, у которых корелляция друг с другом свыше 95 процентов и оставлять среди двух таких один). К счастью, не нашлось подобных признаков не нашлось.

## ElasticNet

### ElasticNet: Объединение силы L1 и L2 регуляризации

ElasticNet - это метод регуляризации в линейной регрессии, который сочетает в себе преимущества L1 (Lasso) и L2 (Ridge) регуляризации. Он был разработан для преодоления ограничений каждого из этих методов по отдельности и обеспечения более стабильного и точного решения в задачах с высокой размерностью и коррелированными признаками.

**Идея ElasticNet**

Основная идея ElasticNet заключается в добавлении к функции потерь линейной регрессии штрафа, который является комбинацией L1 и L2 норм вектора коэффициентов. Это позволяет одновременно:

* **Отбирать признаки (L1):** L1-норма (сумма абсолютных значений коэффициентов) приводит к занулению некоторых коэффициентов, что фактически исключает соответствующие признаки из модели. Это полезно для уменьшения размерности и повышения интерпретируемости модели.
* **Регулировать величину коэффициентов (L2):** L2-норма (сумма квадратов коэффициентов) штрафует большие коэффициенты, что помогает предотвратить переобучение и повысить устойчивость модели к шуму в данных.

**Теоретическая часть**

Рассмотрим стандартную задачу линейной регрессии, где мы хотим предсказать целевую переменную $y$ на основе вектора признаков $x$:

$$y = X\beta + \epsilon$$

где $X$ - матрица признаков, $\beta$ - вектор коэффициентов, а $\epsilon$ - вектор ошибок.

**Функция потерь** для обычной линейной регрессии (метод наименьших квадратов) выглядит следующим образом:

$$L(\beta) = \frac{1}{2n} \sum_{i=1}^{n} (y_i - x_i^T \beta)^2 = \frac{1}{2n} ||y - X\beta||^2$$

**ElasticNet добавляет к этой функции потерь штрафной член:**

$$L_{ElasticNet}(\beta) = \frac{1}{2n} ||y - X\beta||^2 + \lambda (\alpha ||\beta||_1 + \frac{1-\alpha}{2} ||\beta||_2^2)$$

где:

* $\lambda$ - параметр регуляризации, контролирующий силу штрафа.
* $\alpha$ - параметр смешивания, определяющий соотношение между L1 и L2 регуляризацией. При $\alpha=1$ ElasticNet сводится к Lasso, а при $\alpha=0$ - к Ridge.

**Математика**

Для нахождения оптимальных коэффициентов $\beta$ необходимо минимизировать функцию потерь ElasticNet. Это можно сделать с помощью различных методов оптимизации, таких как градиентный спуск или координатный спуск.

**Градиентный спуск:**

Градиент функции потерь ElasticNet по $\beta$ выглядит следующим образом:

$$\nabla L_{ElasticNet}(\beta) = \frac{1}{n} X^T(X\beta - y) + \lambda (\alpha \text{sign}(\beta) + (1-\alpha) \beta)$$

где $\text{sign}(\beta)$ - вектор, элементы которого равны знаку соответствующих элементов $\beta$.

Итеративный процесс обновления коэффициентов $\beta$ с помощью градиентного спуска выглядит так:

$$\beta^{(t+1)} = \beta^{(t)} - \eta \nabla L_{ElasticNet}(\beta^{(t)})$$

где $\eta$ - скорость обучения.

**Координатный спуск:**

Координатный спуск - это итеративный метод, который оптимизирует функцию потерь по одному коэффициенту за раз, фиксируя остальные. Для ElasticNet координатный спуск является более эффективным, чем градиентный спуск, особенно в задачах с высокой размерностью.

**Применение**

ElasticNet широко применяется в различных областях, включая:

* **Анализ геномных данных:** Отбор генов, связанных с определенным заболеванием.
* **Финансовое моделирование:** Прогнозирование цен акций и управление рисками.
* **Обработка изображений:** Восстановление изображений и распознавание объектов.
* **Машинное обучение:** Повышение точности и интерпретируемости моделей.

**Преимущества ElasticNet**

* **Отбор признаков:** Автоматически отбирает наиболее важные признаки, что упрощает модель и повышает ее интерпретируемость.
* **Регуляризация:** Предотвращает переобучение и повышает устойчивость модели к шуму в данных.
* **Работа с коррелированными признаками:** Эффективно справляется с ситуациями, когда признаки сильно коррелированы.
* **Гибкость:** Параметр $\alpha$ позволяет контролировать соотношение между L1 и L2 регуляризацией, что дает возможность адаптировать метод к конкретной задаче.

**Недостатки ElasticNet**

* **Выбор параметров:** Требуется настройка двух параметров: $\lambda$ и $\alpha$, что может быть трудоемким процессом.
* **Вычислительная сложность:** Может быть вычислительно затратным для очень больших наборов данных.

**Сравнение с Lasso и Ridge**

| Метод | Штрафной член | Отбор признаков | Регуляризация | Коррелированные признаки |
|---|---|---|---|---|
| Lasso | $\lambda ||\beta||_1$ | Да | Да | Выбирает один из коррелированных признаков |
| Ridge | $\frac{1}{2} \lambda ||\beta||_2^2$ | Нет | Да | Уменьшает коэффициенты коррелированных признаков, но не зануляет их |
| ElasticNet | $\lambda (\alpha ||\beta||_1 + \frac{1-\alpha}{2} ||\beta||_2^2)$ | Да | Да | Группирует коррелированные признаки и выбирает либо все, либо ни одного |

**Выбор параметров $\lambda$ и $\alpha$**

Для выбора оптимальных значений параметров $\lambda$ и $\alpha$ обычно используются методы кросс-валидации. Данные разбиваются на несколько частей, модель обучается на одних частях и тестируется на других. Значения $\lambda$ и $\alpha$, которые дают наилучшую производительность на тестовых данных, считаются оптимальными.

**Заключение**

ElasticNet - это гибкий метод регуляризации, который сочетает в себе преимущества L1 и L2 регуляризации. Он позволяет эффективно отбирать признаки, регулировать величину коэффициентов и справляться с коррелированными признаками. Благодаря своим преимуществам, ElasticNet широко применяется в различных областях машинного обучения и анализа данных.

###  Обучение модели

Для начала мы масштабируем данные для облегчения работы модели **ElasticNet**

```python
# Масштабирование данных
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Затем определим гиперпараметры для **ElasticNet** и обучим лучшую модель на всех данных c помошью **GridSearch**

```python
# Определение сетки параметров для GridSearchCV
param_grid = {
    'alpha': [0.005, 0.01, 0.1, 1, 10, 100],
    'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]
}

# Создание модели ElasticNet
elastic_net = ElasticNet()

# Создание объекта GridSearchCV
grid_search = GridSearchCV(estimator=elastic_net, param_grid=param_grid, 
                           scoring='neg_mean_squared_error', cv=4, verbose=1)

# Обучение GridSearchCV на тренировочных данных
grid_search.fit(X_train_scaled, y_train)

# Вывод лучших параметров
print("Best parameters found: ", grid_search.best_params_)
print("Best RMSE from GridSearchCV: ", np.sqrt(-grid_search.best_score_))

# Получение лучшей модели
best_elastic_net = grid_search.best_estimator_
```

Сделаем предсказания на тестовой выборке (30% от всех данных) и выведем точность метрикой ***RMSE*** и  ***MAPE***

```python
import numpy as np

def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true))
```

```python
# Прогноз на тестовой выборке
y_pred = best_elastic_net.predict(X_test_scaled)

# Оценка производительности на тестовой выборке
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE on test set:", rmse_test)

# Используем y_test и y_pred из предыдущего примера
mape = calculate_mape(y_test, y_pred)
print("MAPE on test set:", mape)
```

То есть, наша колеблется в районе 11% от значений таргета.
